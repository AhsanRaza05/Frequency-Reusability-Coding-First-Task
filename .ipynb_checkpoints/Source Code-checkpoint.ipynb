# Version # 4

 ### Updated the condition of inner most loop to reduce the number of iterations in worst case scenario.

```python
# Version # 4

import math as ma
import numpy as np

N = input('Please Enter the value of N [For Multiple Seperate by , (comma)]:')

N = N.split(',')

N = np.array(N,dtype = np.int64)

sqrt = 0

for a in np.arange(0,len(N)):
  print("For N:", N[a])

  sqrt = int(ma.sqrt(N[a]))

  if sqrt ** 2 == N[a]:
    print("\t RESULT: i = {i}, j = {j}".format(i = sqrt,j = 0))

  else:
    
    for i in np.arange(1,sqrt + 1):
        
      # Updated the condition of this inner most loop to reduce the number of iterations in worst case scenario.
      for j in np.arange(0,int(ma.sqrt(N[a])) + 1):

        if i ** 2 + i * j + j ** 2 == N[a]:
          print("\t RESULT: i = {i}, j = {j}".format(i = i,j = j))
          break

      else:
          continue

      break

    else:
      print("\t i & j does not exist!")
# Source for Else Statement with For Loop: https://www.geeksforgeeks.org/using-else-conditional-statement-with-for-loop-in-python/
# Source for Numpy Array Creation(i.e: np.arange): https://numpy.org/doc/stable/user/basics.creation.html
# Source: Break Out Outer Loop through inner loop in Python: https://note.nkmk.me/en/python-break-nested-loops/
```

# Version # 2

```python
# Version # 2

import math as ma
import numpy as np

N = input('Please Enter the value of N [N >= 4]:')

N = int(N)

sqrt = 0

print("\nFor N:", N)

sqrt = int(ma.sqrt(N))

if N < 4:
  print("N must be greater than 3!")

else:
  if  sqrt * sqrt == N:
    print("\t RESULT: i = {i}, j = {j}".format(i = sqrt,j = 0))

  else:
    for i in np.arange(1,int(ma.sqrt(N)) + 1):
      for j in np.arange(0,i):

        if i * i + i * j + j * j == N:
          print("\t RESULT: i = {i}, j = {j}".format(i = i,j = j))
          break

      else:
          continue

      break

    else:
      print("\t i & j does not exist!")


# Source for Else Statement with For Loop: https://www.geeksforgeeks.org/using-else-conditional-statement-with-for-loop-in-python/
# Source for Numpy Array Creation(i.e: np.arange): https://numpy.org/doc/stable/user/basics.creation.html
# Source: Break Out Outer Loop through inner loop in Python: https://note.nkmk.me/en/python-break-nested-loops/
```
