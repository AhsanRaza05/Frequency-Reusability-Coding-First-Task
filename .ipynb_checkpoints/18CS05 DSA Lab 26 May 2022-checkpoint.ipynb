```python
import matplotlib.pyplot as py

# matplotlib > Library
# pyplot is Sub Module of matplotlib
```

```python


classes = ["First", "Second", "Third"]
Positions = [1, 2, 3]

py.plot(classes,Positions)
```

```python
py.scatter(classes,Positions)
```

```python
x = range(1,21)
y = []
for i in x:
    
    y.append(i + 1)
    
py.plot(x,y)
py.title("Plot Linear Line")
py.xlabel("X values")
py.ylabel("Y values")
py.show()
```

```python
x = range(1,21)
y = []
for i in x:
    
    y.append(i + 1)
    
py.plot(x,y)
py.grid()
py.title("Plot Linear Line")
py.xlabel("X values")
py.ylabel("Y values")
py.show()
```

```python
classes = ["First", "Second", "Third"]
Positions = [1, 2, 3]

py.plot(classes,Positions, marker = "+", color ='tab:orange')
```
